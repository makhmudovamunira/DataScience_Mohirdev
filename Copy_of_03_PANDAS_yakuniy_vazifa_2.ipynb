### 7-VAZIFA. AQSH gidro elektr stansiyalardan eng kam va eng ko'p elektr energiyasi olgan yillarni toping.

```python
# JAVOBNI SHUYERGA YOZING
maxyil=dfus.loc[:,'komir':'atom'].sum(axis=1).idxmax()
minyil=dfus.loc[:,'komir':'atom'].sum(axis=1).idxmin()
print(f"AQSH {maxyil}-yilda eng ko'p elektr energiyasi ishlab chiqqan\n {minyil}-yilda eng kam elektr energiyasini ishlab chiqqan")
```
