```python
df1=pd.DataFrame(np.arange(9.).reshape=(3,3), columns=list('bcd'), index=['Olma', 'Anor','Uzum'])
df2=pd.DataFrame(np.arange(12.).reshape=(3,4), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.add(df2)
```

```python
df1-df2
```

```python
df1.add(df2, fill_value=0) # NaN bo'lgan qiymat o'rniga 0 qo'yadi
```

```python
df1.sub(df2, fill_value=0) # df1-df2
```

```python
df1.rsub(df2) # df2-df1 degan manoni beradi
```

**Pandas elementlari obyektlariga funksiyalarni qo'llash**

```python
df = pd.DataFrame(np.random.randn(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
np.abs(df)
```

```python
np.round(df)
```

```python
df.mean()
```

```python
df.apply(np.mean) #meanni apply orqali qo'llash
```

Funksiya nafaqat yagona qiymat, balki series obyektini ham qaytarishi mumkin

```python
def foo(x):
  return pd.Series([x.min(),x.max()], index=['min','max'])

df.apply(foo)
```
